```python
# Context:
# https://twitter.com/_WTProject/status/1172916785764585473
# https://twitter.com/LargeCardinal/status/1173167299060215809

%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt


def rot(xy, deg):
    rad = deg * np.pi / 180
    R = np.array([
            [np.cos(rad), -np.sin(rad)],
            [np.sin(rad),  np.cos(rad)],
        ])
    return R.dot(xy)
    
def f(deg):
    A = np.array([162, 0])
    B = rot([72, 0], 180-deg) + np.array([108, 0])
    return np.linalg.norm(A - B)

X = np.arange(180)
Y = np.array([np.vectorize(f)(X)]).T

plt.figure(figsize=(9,5))
plt.plot(X, Y)
plt.plot([0, 100], [0, 100])
plt.xlabel("Angle (degrees)")
plt.ylabel("Right side (unit lengths)")
plt.grid()
```
