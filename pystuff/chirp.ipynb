```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

N = 300
maxfreq = np.pi
oversampling = 10

X = np.arange(0, N, 1.0 / oversampling)

Y = list()
state = 1.0 + 0.0j
for i in X:
    Y.append(np.real(state))
    state *= np.exp(1j*maxfreq*i/N/oversampling)

plt.figure(figsize=(10, 3))
plt.title("Oversampled Waveform")
plt.plot(X, Y)
plt.show()

plt.figure(figsize=(10, 3))
plt.title("Without Oversampling")
plt.plot(Y[0::oversampling])
plt.show()
```
