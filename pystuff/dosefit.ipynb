```python
%matplotlib inline
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

xdata=np.array([20, 40, 60, 50])
ydata=np.array([8, 25, 155, 91])

def model_1(X, a, b, c):
    return [a + b*np.exp(c*x) for x in X]

def model_2(X, a, b):
    return [a*np.exp(b*x)-a for x in X]

popt_1, _ = curve_fit(model_1, xdata, ydata, [0.1, 0.1, 0.1])
popt_2, _ = curve_fit(model_2, xdata, ydata, [0.1, 0.1])
print()

print("Parameters for Model 1 (A + B*exp(C*x)):")
print(popt_1)
print()

print("Parameters for Model 2 (A*exp(B*x) - A):")
print(popt_2)
print()

X = np.linspace(0, 180, 1000)
Y1 = model_1(X, *popt_1)
Y2 = model_2(X, *popt_2)

plt.figure(figsize=(5, 5))
plt.plot(xdata, ydata, 'o')
plt.plot(X, Y1, label="model_1")
plt.plot(X, Y2, label="model_2")
plt.xlim([0, 80])
plt.ylim([0, 180])
plt.legend(loc='upper left');
```
