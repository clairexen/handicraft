```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from urllib.request import urlopen
from scipy.optimize import curve_fit
from typing import Iterable
import json, os, ssl
```

```python
sslctx = ssl.SSLContext(ssl.PROTOCOL_SSLv23)
sslctx.set_ciphers('ALL@SECLEVEL=1')
txtdata = urlopen('https://info.gesundheitsministerium.at/data/Epikurve.js', context=sslctx).read().decode()
# print(txtdata, end="\n\n")
jsdata = json.loads(txtdata.split("\n")[0].split(" = ")[1][:-1])
# print(jsdata, end="\n\n")
assert jsdata[0]['label'] == '25.02.2020'
jsdata.insert(0, {"label": '24.02.2020', "y": 0})
jsdata.insert(0, {"label": '23.02.2020', "y": 0})
print(jsdata, end="\n\n")

data, dates = zip(*[(d["y"], ".".join(d["label"].split(".")[0:2])) for d in jsdata])
data, dates = list(data), list(dates)
print(np.array(data), end="\n\n")
data = np.cumsum(data)
print(data, end="\n\n")
print(dates, end="\n\n")
print(data[-1], dates[-1], len(data))
```

```python
for nweeks,ylim in [(15, 1000), (-1, 1200)]:
    fig = plt.figure(figsize=(18,6))
    fig.patch.set_facecolor('white')

    plt.ylabel("new covid-19 cases")
    plt.bar(np.arange(len(data)-1)+0.5, np.diff(data), color='orange')

    plt.xticks(range((len(data)+6)%7, len(data), 7), dates[(len(data)+6)%7::7])
    if nweeks < 0:
        plt.xlim(0, len(data)-1)
    else:
        plt.xlim(len(data)-(7*nweeks+1), len(data)-1)
    plt.ylim(0, ylim)

    if True:
        X, Y = [], []
        if nweeks < 0:
            r = range(10, len(data))
        else:
            r = range(len(data)-(7*nweeks+1), len(data))
        for i in r:
            X.append(i-0.5)
            Y.append(np.mean(np.diff(data[i-7:i+1])))
        plt.plot(X, Y, "o")       

    if True:
        if nweeks < 0:
            n = len(data)//7-4
        else:
            n = nweeks

        for i in range(n):
            x1 = len(data)-8-i*7
            x2 = len(data)-1-i*7
            y = (data[x2] - data[x1]) / 7
            plt.plot([x1, x2], [y, y], "r-")

    plt.title("covid-19 infections in Austria with weekly averages\n"
              "(data from BMSGPK dataset via info.gesundheitsministerium.at)")
    plt.grid()
```

```python
nweeks = 10
for N, Nstr in [
    (2, "two"),
    # (3, "three"),
    (5, "five"),
]:
    avgdata = np.array([np.sum(np.diff(data)[i:i+N])/N for i in range(len(data)-N)])
    k = len(avgdata)

    if True:
        fig = plt.figure(figsize=(18,4))
        fig.patch.set_facecolor('white')

        plt.plot(avgdata, "y")
        plt.plot(avgdata, "ok")

        plt.ylim(0, 800)
        plt.xlim(0, k-1)
        plt.xticks(np.arange((k-1)%7, k, 7))
        plt.title("daily new covid-19 infections in Austria over %s-day averages\n"
                  "(data from BMSGPK dataset via info.gesundheitsministerium.at)" % Nstr)
        plt.grid()

    if True:
        fig = plt.figure(figsize=(18,4))
        fig.patch.set_facecolor('white')

        for d in range(7):
            daydata = avgdata[k-7*nweeks-d-1:k-d:7]
            plt.plot(np.arange(70-nweeks-10*d, 70-10*d+1), daydata)

        plt.ylim(0, 800)
        plt.xlim(0, 70)
        plt.title("weekday trend for covid-19 infections in Austria over %s-day averages\n"
                  "(data from BMSGPK dataset via info.gesundheitsministerium.at)" % Nstr)
        plt.grid()
    
    if False:
        fig = plt.figure(figsize=(18,4))
        fig.patch.set_facecolor('white')

        for d in range(7):
            daydata = avgdata[k-7*nweeks-d-1:k-d:7]
            plt.plot(np.arange(k-7*nweeks-d-1, k-d, 7), daydata)

        plt.plot(avgdata, "k:")

        plt.ylim(0, 160)
        plt.xlim(k-7*nweeks-7, k)
        plt.title("weekday trend for covid-19 infections in Austria over %s-day averages\n"
                  "(data from BMSGPK dataset via info.gesundheitsministerium.at)" % Nstr)
        plt.grid()
```

```python
def expit(x):
    return 1/(1+np.exp(-x))

def intexpit(x):
    return np.log(np.exp(x)+1)

def f1(x, xs, xo, ys):
    return f2(x, xs, xo, ys, 0)

def f2(x, xs, xo, ys, k):
    if isinstance(x, Iterable):
        return [f2(v, xs, xo, ys, k) for v in x]
    else:
        y = ys*expit(xs*x+xo)
        y += k*intexpit(xs*x+xo)
        return y

def plotit(horizon):
    p1_0 = [0.22, -7.0, 14000.0]
    p2_0 = [0.26, -7.7, 10000.0, 640.0]

    xd = np.arange(horizon)
    yd = data[0:horizon]

    p1, pcov1 = curve_fit(f1, xd, yd, p1_0)
    #print(p1)
    #print(pcov1)

    p2, pcov2 = curve_fit(f2, xd, yd, p2_0)
    #print(p2)
    #print(pcov2)

    plt.title(("non-linear lsq fit for logistic function [+ linear tail], fit for %d samples\n" +
               "*** THIS MODEL IS GARBAGE *** USE FOR ENTERTAINMENT PURPOSES ONLY ***") % horizon)
    plt.ylabel("total confirmed covid-19 cases in Austria")
    plt.xlabel("days since %s" % dates[0])
    plt.plot(data, "o")
    plt.plot(np.arange(0, 201), f1(np.arange(0, 201), *p1), "--")
    plt.plot(np.arange(0, 201), f2(np.arange(0, 201), *p2), "-.")
    plt.plot([horizon, horizon], [0, 24000], "r")
    plt.yticks(range(0, 40001, 4000))
    plt.ylim(0, 40000)
    plt.xlim(0, 250)
    plt.grid()

if False:
    for horizon in range(24, len(data)):
        print("Generating frame for hoizon=%d.." % horizon)
        fig = plt.figure(figsize=(14,4))
        fig.patch.set_facecolor('white')
        plotit(horizon)
        plt.savefig("model_%03d.png" % horizon)
        plt.close()
    print("Generating GIF animation..")
    os.system("convert model_*.png -coalesce -loop 0 model_anim.gif")
    print("done.")
else:
    for horizon in (30, 40, 60, 80):
        fig = plt.figure(figsize=(14,4))
        fig.patch.set_facecolor('white')
        plotit(horizon)
```
