```python
%pylab inline
import tensorflow as tf
from IPython.display import clear_output
db = np.loadtxt('testdata.txt')
```

```python
figure(figsize=(16,8))
for i in range(16):
    subplot(4, 4, i)
    plot(db[i, 0:30], 'k')
    plot(10*db[i, 30:60], 'b.-')
```

```python
# ANN Model

# Input Placeholder
x_n = 9
x = tf.placeholder("float", [None, x_n])

# Input Layer (Sigmoid)
L1_n = 16
L1_W = tf.Variable(tf.truncated_normal((x_n, L1_n), stddev=0.1))
L1_b = tf.Variable(tf.truncated_normal((L1_n,), stddev=0.1))
L1 = tf.nn.sigmoid(tf.matmul(x, L1_W) + L1_b)

# First Hidden Layer
L2_n = 32
if False: # RELU
    L2_W = tf.Variable(tf.truncated_normal((L1_n, L2_n), stddev=0.1))
    L2_b = tf.Variable(tf.constant(0.1, shape=(L2_n,)))
    L2 = tf.nn.relu(tf.matmul(L1, L2_W) + L2_b)
else: # Sigmoid
    L2_W = tf.Variable(tf.truncated_normal((L1_n, L2_n), stddev=0.1))
    L2_b = tf.Variable(tf.truncated_normal((L2_n,), stddev=0.1))
    L2 = tf.nn.sigmoid(tf.matmul(L1, L2_W) + L2_b)

# Second Hidden Layer
L3_n = 32
if False: # RELU
    L3_W = tf.Variable(tf.truncated_normal((L2_n, L3_n), stddev=0.1))
    L3_b = tf.Variable(tf.constant(0.1, shape=(L3_n,)))
    L3 = tf.nn.relu(tf.matmul(L2, L3_W) + L3_b)
else: # Sigmoid
    L3_W = tf.Variable(tf.truncated_normal((L2_n, L3_n), stddev=0.1))
    L3_b = tf.Variable(tf.truncated_normal((L3_n,), stddev=0.1))
    L3 = tf.nn.sigmoid(tf.matmul(L2, L3_W) + L3_b)

# Dropout Layer
L4_n = L3_n
L4_p = tf.placeholder("float")
L4 = tf.nn.dropout(L3, L4_p)

# Output Layer (Softmax)
L5_n = 4
L5_W = tf.Variable(tf.truncated_normal((L4_n, L5_n), stddev=0.1))
L5_b = tf.Variable(tf.constant(0.1, shape=(L5_n,)))
L5 = tf.nn.softmax(tf.matmul(L4, L5_W) + L5_b)

# Output Placeholder
y = L5
y_ = tf.placeholder("float", [None, L5_n])
cross_entropy = -tf.reduce_sum(y_*tf.log(y))
train_step = tf.train.AdamOptimizer(1e-4).minimize(cross_entropy)
```

```python
def next_batch(n):
    xs = []; ys = []
    for i in range(n):
        k = np.random.randint(db.shape[0])
        p = np.random.randint(5, 25)
        while (i%2 == 0) != (db[k, 30+p] < 0.5):
            p = np.random.randint(5, 25)
        xs.append(db[k, p-4:p+5] / np.max(db[k, p-4:p+5]))
        ys.append([db[k, 30+p-1], db[k, 30+p], db[k, 30+p+1], 1-np.sum(db[k, 30+p-1:30+p+2])])
    return xs, ys

batch_xs, batch_ys = next_batch(16)
figure(figsize=(16,8))
for i in range(16):
    subplot(4, 4, i)
    plot(batch_xs[i], 'k')
    plot([3, 4, 5], batch_ys[i][0:3], 'b.-')

sess = tf.Session()
sess.run(tf.initialize_all_variables())
train_iter = 0
```

```python
for cnt in range(100):
    for i in range(500):
        batch_xs, batch_ys = next_batch(100)
        sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys, L4_p: 0.5})
    
    clear_output()
    
    train_iter += 1
    print("At iteration %d:" % train_iter)
    output_ys = sess.run(y, feed_dict={x: batch_xs, L4_p: 1.0})

    figure(figsize=(16, 6))
    for i in range(20):
        subplot(4, 5, i)
        plot(batch_xs[i], 'k')
        plot([3, 4, 5], batch_ys[i][0:3], 'b.-')
        plot([3, 4, 5], output_ys[i][0:3], 'r.-')
    
    show()

print("DONE.")
```
