```python
%matplotlib inline
import numpy as np
from math import gcd
import matplotlib.pyplot as plt
```

```python
def genpies(with_bug=False):
    count_coprime = 0
    count_cofactor = 0
    pies = list()

    for i in range(100000):
        a = np.random.randint(1, 121)
        b = np.random.randint(1, 121)
        
        is_coprime = gcd(a, b) == 1
        
        if with_bug and (a == 1 or b == 1):
            is_coprime = False

        if is_coprime:
            count_coprime += 1
        else:
            count_cofactor += 1

        x = count_coprime / (count_coprime + count_cofactor)
        if x != 0:
            pies.append(np.sqrt(6 / x))
        else:
            pies.append(0)
            
    return pies

np.random.seed(1234)

plt.figure(figsize=(16, 4))
plt.plot(genpies(with_bug=True), label="algo with bug")
plt.plot(genpies(with_bug=False), label="algo without bug")
plt.plot([0, 100000], [np.pi, np.pi], label="pi")
plt.ylim(3.1, 3.3)
plt.legend()
plt.show()
```
