```python
%pylab inline
from __future__ import division
from __future__ import print_function
```

```python
data=[]
with file('azone_01.txt', 'r') as f:
    for line in f:
        d = line.replace(':', ' ').split()
        data.append(int(d[0])*60*60+int(d[1])*60+int(d[2]))

data = np.array(data)
timesteps = diff(data)

p = np.poly1d(np.polyfit(arange(len(timesteps)), timesteps, 2))
model = p(arange(1000))
partial_sums = np.cumsum(model)

figure(figsize=(10,5))
plot(timesteps, '.', label='measured timesteps')
plot(model[0:len(timesteps)], label='model fit')
xlabel('proven steps in k-seq')
ylabel('seconds per step')
legend(loc=2)
grid()

figure(figsize=(10,5))
plot(timesteps, '.', label='measured timesteps')
plot(model, label='model extrapolation')
xlabel('proven steps in k-seq')
ylabel('seconds per step')
legend(loc=2)
grid()

figure(figsize=(10,5))
plot(data / (60*60), label='measured time')
plot(partial_sums[0:len(data)] / (60*60), label='model')
xlabel('proven steps in k-seq')
ylabel('total hours spent')
legend(loc=2)
grid()

figure(figsize=(10,5))
plot(data / (60*60), label='measured time')
plot(partial_sums / (60*60), label='model')
xlabel('proven steps in k-seq')
ylabel('total hours spent')
legend(loc=2)
grid()
```
