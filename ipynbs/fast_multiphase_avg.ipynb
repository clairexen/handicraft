This notebook demonstrates how to efficiently create a bandwith-limited model for a signal from a set of unaligned and undersampled oberservations. See the function `fast_multiphase_avg()` below for an efficient implementation of the algorithm.

First we create a $M$ noisy observations with $N$ samples each. The phase for those observations is chosen randomly.

```python
M = 200
N = 16

phase = np.random.uniform(-1, +1, M)
samples = np.zeros((M, N))

for i in range(M):
    samples[i, :] = 25 * np.exp(-((np.arange(N)-N*0.7+phase[i])*0.5)**2) * (1 + np.cos(5*(np.arange(N)+phase[i]))) + \
                    50 * np.exp(-(np.arange(N)-N/4+phase[i])**2) + np.random.normal(size=N)
```

Before we continue lets plot the observed x/y tuples.

```python
points_x = np.zeros((M, N))
points_y = np.zeros((M, N))

for i in range(M):
    points_y[i, :] = samples[i, :]
    points_x[i, :] = np.arange(N) + phase[i]

figure(figsize=(14,5))
plot(points_x.flat, points_y.flat, ',');
```

The algorithm is using $P$ buckets. Each observation is assigned a bucket based on the phase (simmilar phases are assigned to the same bucket).

```python
P = 8
buckets = [list() for i in range(P)]

for i in range(M):
    p = int(round(P * phase[i]) % P)
    buckets[p].append(i)

print([len(b) for b in buckets])

figure(figsize=(14, 5))
for p in range(P):
    points_x = list()
    points_y = list()
    for i in buckets[p]:
        for n in range(N):
            points_x.append(n + phase[i])
            points_y.append(samples[i, n])
    plot(points_x, points_y, '.', label="bucket %d" % p)
legend();
```

Create a low resolution FFT for each observation. Correct the phase of each observation in the frequency domain to align with the center of the bucket. Then average all FFTs per bucket to create one low-res FFT for each bucket, and then extract samples with iFFT (plus phase shift). Interlace the samples from the individual buckets and finally lowpass filter the results.

```python
# Fine Samples
FS = np.zeros(N*P)

# Plot each phase bucket
figure(figsize=(14, 14))
subplot(4, 2, 1)

for p in range(P):
    # avg. DFT of this bucket (after phase correction for each shot)
    
    sum_dft = np.zeros(N/2+1, np.complex)
    
    for i in buckets[p]:
        dft = np.fft.rfft(samples[i, :])
        for k in range(len(dft)):
            dft[k] *= exp(-2j * pi * phase[i] * k / N)
        sum_dft += dft
    
    sum_dft /= len(buckets[p])
    
    # Extract fine samples
    
    phase_dft = np.copy(sum_dft)
    for k in range(len(phase_dft)):
        phase_dft[k] *= exp(2j * pi * p * k / N / P)
    FS[p::P] = np.fft.irfft(phase_dft)
    
    # Plot this bucket
    
    sum_dft *= P
    sum_dft[-1] *= 0.5
    sum_dft.resize(P*N/2+1)

    subplot(4, 2, p+1)
    
    for q in range(P):
        if q != p:
            for i in buckets[q]:
                plot(np.arange(N) + phase[i], samples[i, :], 'y.')

    for i in buckets[p]:
        plot(np.arange(N) + phase[i], samples[i, :], 'r.')
        
    plot(np.arange(P*N)/P, np.fft.irfft(sum_dft), 'k-')

# Lowpass filter fine samples

dft = np.fft.rfft(FS)
dft *= np.abs(np.fft.rfftfreq(len(FS))*2*P) < 3
FS = np.fft.irfft(dft)

# Plot fine samples

figure(figsize=(14, 5))
for i in range(M):
    plot(np.arange(N) + phase[i], samples[i, :], 'y.')

plot(np.arange(P*N)/P, FS, 'k.');
```

Now lets do the same thing, but much faster. Only perform one FFT per input waveform, do the entire algorithm in the frequency domain (using Radix-2 Cooley–Tukey Butterflies for interlacing the individual buckets), and only run one iFFT at the end. **Copy the following cell into your code or use `import fast_multiphase_avg` (using `fast_multiphase_avg.py`) if you need this functionality.**

```python
def butterfly_combine_dft(dft_even, dft_odd):
    """Combine two N/2-DFTs into one N-DFT using a Radix-2 Cooley–Tukey Butterfly.
    
    Combine the FFT of even (0, 2, 4, ...) and odd (1, 3, 5, ...) samples of a signal,
    as returned by np.fft.fft(), into the FFT for the entire signal.
    """
    
    N = len(dft_even)
    assert N == len(dft_odd)
    dft = np.zeros(2 * N, np.complex)
    omegas = np.exp(-2j * np.pi * np.arange(N) / (2*N))
    dft[0:N] = dft_even + dft_odd*omegas
    dft[N:2*N] = dft_even - dft_odd*omegas
    return dft   

def fast_multiphase_avg(samples, phases, oversampling, bwlimit):
    """Average multiple undersampled waveforms at arbitrary phases into one oversampled waveform.
    
    This function combines undersampled, noisy waveforms taken at arbitrary phases, into one
    oversampled waveform.
    
    Parameters
    ----------
    samples: 2d array
        1st dimension: number of waveforms
        2nd dimension: number of samples per waveform
    phases: 1d array
        phase for each waveform. 1 = offset by one sample
    oversampling: int, power of two
        oversampling factor
    bwlimit: real
        bandwith limit of the signal, given as factor to
        the nyquist frequency of the undersampled waveforms
    """
    assert bin(oversampling).count("1") == 1
    assert samples.shape[0] == len(phases)
    
    # many forward DFTs
    samples_dft = np.fft.fft(samples)
    freq_dft = np.fft.fftfreq(samples.shape[1])
    
    # correct phases and assign buckets
    buckets = [list() for i in range(oversampling)]
    for i in range(samples_dft.shape[0]):
        b = int(round(phases[i] * oversampling)) % oversampling
        p = phases[i] - b / oversampling
        samples_dft[i, :] *= np.exp(-2j * np.pi * p * freq_dft)
        buckets[b].append(i)
    
    # create per-bucket avg. FFTs
    avg_dft = np.zeros((oversampling, samples_dft.shape[1]), np.complex)
    for b in range(oversampling):
        avg_dft[b, :] = np.mean(samples_dft[buckets[b], :], 0)
    
    # combine individual FFTs using butterflies
    while avg_dft.shape[0] > 1:
        new_avg_dft = np.zeros((avg_dft.shape[0] // 2, avg_dft.shape[1] * 2), np.complex)
        for i in range(new_avg_dft.shape[0]):
            new_avg_dft[i, :] = butterfly_combine_dft(avg_dft[i, :], avg_dft[i + new_avg_dft.shape[0], :])
        avg_dft = new_avg_dft
    
    # low-pass filter results
    avg_dft[0, :] *= np.abs(np.fft.fftfreq(avg_dft.shape[1]))*2*oversampling < bwlimit

    # obtain oversampled time-domain representation using an inverse FFT
    results_x = np.arange(avg_dft.shape[1]) / oversampling
    results_y = np.real(np.fft.ifft(avg_dft[0, :]))
    
    return results_x, results_y
```

Finally test this implementation:

```python
fine_x, fine_y = fast_multiphase_avg(samples, phase, 8, 3)

figure(figsize=(14, 5))
for i in range(M):
    plot(np.arange(N) + phase[i], samples[i, :], 'y.')

plot(fine_x, fine_y, 'k.');
```
