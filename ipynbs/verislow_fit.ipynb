```python
%pylab inline
X = [1, 2, 4, 8, 16, 32, 48, 64]
Y = [0.9, 1.48, 4.14, 12.94, 45.26, 170.63, 397.91, 765.96]
plot(X, Y, 'o:')
xlabel('#k words of memory')
ylabel('Elaboration time in seconds')
```

```python
from scipy.optimize import curve_fit
f = lambda x, p1, p2, p3: p1*x + p2*x**p3
popt, pcov = curve_fit(f, X, Y)
print(popt)

plot(X, Y, 'o')
plot(np.arange(70), f(np.arange(70), *popt))
xlabel('#k words of memory')
ylabel('Elaboration time in seconds')
```
