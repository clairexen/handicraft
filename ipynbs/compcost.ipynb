```python
# data from https://en.wikipedia.org/wiki/FLOPS#Hardware_costs

data = [
    (1961, 8300000000000.0),
    (1984, 42780000.0),
    (1997, 42000.0),
    (2000, 1300.0),
    (2000, 836.0),
    (2003, 100.0),
    (2007, 52.0),
    (2011, 1.80),
    (2013, 0.22),
    (2013, 0.16),
    (2013, 0.12),
    (2015, 0.08)
]
```

```python
# Does it look like an exponential trend? It sure does..
plot([x for x,y in data], [log10(y) for x,y in data], "o")
```

```python
# fit a linear function to the log plot
X = np.array([x for x,y in data])
Y = np.array([log10(y) for x,y in data])

M = vstack([X, ones(X.shape)])
p = solve(M.dot(M.transpose()), M.dot(Y))

def f(x):
    return p[0]*x + p[1]

plot([x for x,y in data], [log10(y) for x,y in data], "o")
plot(X, [f(x) for x in X])
```

```python
# Decrease in cost every ten years
pow(10, f(2000) - f(2010))
```

```python
# Decrease in cost every 18 months
pow(10, f(2000) - f(2000+1.5))
```

```python
# Let's make a nice plot
figure(figsize=(9,6))
ax = plt.subplot(111)
ax.semilogy([x for x,y in data], [y for x,y in data], "o")
ax.semilogy(X, [pow(10, f(x)) for x in X])
ax.set_ylabel("Cost of Computing in USD per GFLOP (in today's money)")
ax.set_title("Cost of Computing decreases 450x each decade,\n" +
        "or 2.5x every 18 months: Faster than Moore's law")
ax.text(0.03, 0.90, "1961: 1 GFLOP = 17 Million IBM 1620\n" +
        "         8.3 trillion USD / GFLOP", transform=ax.transAxes)
ax.text(0.41, 0.58, "1984: Cray X-MP/48 at 0.8 GFLOP\n" +
        "     42,780,000 USD / GFLOP", transform=ax.transAxes)
ax.text(0.49, 0.02, " " * 25 + "0.22 USD / GFLOP\n" +
        "2013: Sony PlayStation 4 at 1.84 TFLOPs", transform=ax.transAxes)
ax.text(0.03, 0.02, "Data points from Wikipedia:\n" +
        "https://en.wikipedia.org/wiki/FLOPS#Hardware_costs\n\n" +
        "IPython Notebook by Clifford Wolf", transform=ax.transAxes)
plt.show()
```
