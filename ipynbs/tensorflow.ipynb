```python
# This is essentially the TensorFlow MNIST Tutorial
# http://www.tensorflow.org/tutorials/mnist/pros/

%pylab inline
import mnist_data
import tensorflow as tf
from IPython import display
mnist = mnist_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
# Model (Softmax Regression)
x = tf.placeholder("float", [None, 28*28])   # input: pixels
W = tf.Variable(tf.zeros([28*28,10]))        # variable: weights
b = tf.Variable(tf.zeros([10]))              # variable: biases
y = tf.nn.softmax(tf.matmul(x,W) + b)        # output: predictions
```

```python
# Training
y_ = tf.placeholder("float", [None,10])      # input: labels
cross_entropy = -tf.reduce_sum(y_*tf.log(y))
train_step = tf.train.GradientDescentOptimizer(0.01).minimize(cross_entropy)
```

```python
# Run Training Session
sess = tf.Session()
sess.run(tf.initialize_all_variables())

for i in range(1000):
    batch_xs, batch_ys = mnist.train.next_batch(100)
    sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
```

```python
# Evaluate Model
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
print sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels})
```

```python
# Model (Multilayer Convolutional Network)

def weight_variable(shape):
    initial = tf.truncated_normal(shape, stddev=0.1)
    return tf.Variable(initial)

def bias_variable(shape):
    initial = tf.constant(0.1, shape=shape)
    return tf.Variable(initial)

def conv2d(x, W):
    return tf.nn.conv2d(x, W, strides=[1, 1, 1, 1], padding='SAME')

def max_pool_2x2(x):
    return tf.nn.max_pool(x, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='SAME')

# First Layer: 28x28x1 -> 5x5 convolution -> 14x14x32
W_conv1 = weight_variable([5, 5, 1, 32])
b_conv1 = bias_variable([32])
x_image = tf.reshape(x, [-1,28,28,1])
h_conv1 = tf.nn.relu(conv2d(x_image, W_conv1) + b_conv1)
h_pool1 = max_pool_2x2(h_conv1)

# Second Layer: 14x14x32 -> 5x5 convolution -> 7x7x64
W_conv2 = weight_variable([5, 5, 32, 64])
b_conv2 = bias_variable([64])
h_conv2 = tf.nn.relu(conv2d(h_pool1, W_conv2) + b_conv2)
h_pool2 = max_pool_2x2(h_conv2)

# Third Layer: 7x7x64 -> dense ReLU -> 1024
W_fc1 = weight_variable([7 * 7 * 64, 1024])
b_fc1 = bias_variable([1024])
h_pool2_flat = tf.reshape(h_pool2, [-1, 7*7*64])
h_fc1 = tf.nn.relu(tf.matmul(h_pool2_flat, W_fc1) + b_fc1)

# Dropout
keep_prob = tf.placeholder("float")
h_fc1_drop = tf.nn.dropout(h_fc1, keep_prob)

# Readout (Softmax Regression)
W_fc2 = weight_variable([1024, 10])
b_fc2 = bias_variable([10])
y_conv = tf.nn.softmax(tf.matmul(h_fc1_drop, W_fc2) + b_fc2)
```

```python
# Training, Display accuracy in training data over time
cross_entropy = -tf.reduce_sum(y_*tf.log(y_conv))
train_step = tf.train.AdamOptimizer(1e-4).minimize(cross_entropy)

correct_prediction = tf.equal(tf.argmax(y_conv,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))

sess.run(tf.initialize_all_variables())

train_accuracy_x = []
train_accuracy_y1 = []
train_accuracy_y2 = []

for i in range(20000):
    batch_xs, batch_ys = mnist.train.next_batch(100)
    if i % 10 == 0:
        train_accuracy_x.append(i)
        train_accuracy_y1.append(sess.run(accuracy, feed_dict={x: batch_xs, y_: batch_ys, keep_prob: 0.5}))
        train_accuracy_y2.append(sess.run(accuracy, feed_dict={x: batch_xs, y_: batch_ys, keep_prob: 1.0}))
        # print("step %d, training accuracy %g" % (i, train_accuracy))
        display.clear_output(wait=True)
        figure(figsize=(10, 5))
        plot(train_accuracy_x, train_accuracy_y1, 'r', label="with dropout")
        plot(train_accuracy_x, train_accuracy_y2, 'g', label="without dropout")
        legend(loc='lower right')
        show()
    sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys, keep_prob: 0.5})
```

```python
# Evaluate Model
print sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels, keep_prob: 1.0})

# Show Failures
figure(figsize=(10,5))
p = sess.run(correct_prediction, feed_dict={x: mnist.test.images, y_: mnist.test.labels, keep_prob: 1.0})
for n, idx in enumerate(np.random.permutation(find(list(~p)))[:50]):
    subplot(5, 10, n)
    imshow(mnist.test.images[idx,:].reshape([28,-1]), 'gray')
```
