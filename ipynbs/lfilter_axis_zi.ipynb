```python
%pylab inline --no-import-all
from scipy.signal import cheby1, lfilter, lfiltic

enable_zi_axis_work_around = True

signals = 1.2 + np.zeros((10, 1000))
for i in range(10):
    signals[i, 100+80*i:120+80*i] -= 1.0

plt.figure(figsize=(9, 9))

plt.subplot(3, 1, 1)
plt.title("Unfiltered signals")
plt.plot(signals.T)
plt.ylim(0, 1.5)

b, a = cheby1(3, 0.5, [0.3, 0.7], btype='bandpass')
zi = lfiltic(b, a, [], [1.2] * (len(b)-1))

filtered = lfilter(b, a, signals, axis=1)

plt.subplot(3, 1, 2)
plt.title("Filtered signals (without zi)")
plt.plot(filtered.T)
plt.ylim(-0.5, +0.5)

if enable_zi_axis_work_around:
    for i in range(10):
        filtered[i,:], _ = lfilter(b, a, signals[i,:], zi=zi)
else:
    # it seems like "zi" and "axis" are mutually exclusive
    filtered, _ = lfilter(b, a, signals, axis=1, zi=zi)
        
plt.subplot(3, 1, 3)
plt.title("Filtered signals (with zi)")
plt.plot(filtered.T)
plt.ylim(-0.5, +0.5);
```
