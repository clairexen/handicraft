```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import fpuns

values = fpuns.genPositiveValues(16)
```

```python
for scale in [10000000000, 1000, 10]:
    plt.figure(figsize=(14,4))
    plt.semilogx(values, np.arange(0, len(values)))
    plt.xlim(1/scale, scale)
    plt.ylim(0, len(values))
```
